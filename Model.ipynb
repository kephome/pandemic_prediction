грузим библиотеки

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
```

грузим датасет

```python
dataset = pd.read_csv('covid_data_train.csv')
```

обработка датасета

```python
data = dataset[['volume_serv_medicine_2017', 'urban', 'density', 'humidity_median', 'volume_serv_transport_2017',
                'inf_rate', 'urban_50-54_years', 'urban_55-59_years', 'urban_60-64_years', 'urban_65-69_years',
                'urban_70-74_years', 'urban_75-79_years', 'urban_80-84_years', 'urban_85-89_years', 'urban_90-94_years',
                'rural_50-54_years', 'rural_55-59_years', 'rural_60-64_years', 'rural_65-69_years', 'rural_70-74_years',
                'rural_75-79_years', 'rural_80-84_years', 'rural_85-89_years', 'rural_90-94_years']]
data = data.dropna(axis=0)
data = data.drop_duplicates(ignore_index=True)
data
```

создание своих признаков

```python
data['urban_50+'] = data['urban_50-54_years'] + \
                    data['urban_55-59_years'] + \
                    data['urban_60-64_years'] + \
                    data['urban_65-69_years'] + \
                    data['urban_70-74_years'] + \
                    data['urban_75-79_years'] + \
                    data['urban_80-84_years'] + \
                    data['urban_85-89_years'] + \
                    data['urban_90-94_years']
data = data.drop(['urban_50-54_years', 'urban_55-59_years', 'urban_60-64_years',
                  'urban_65-69_years', 'urban_70-74_years', 'urban_75-79_years',
                  'urban_80-84_years', 'urban_85-89_years', 'urban_90-94_years'], axis=1)

data['rural_50+'] = data['rural_50-54_years'] + \
                    data['rural_55-59_years'] + \
                    data['rural_60-64_years'] + \
                    data['rural_65-69_years'] + \
                    data['rural_70-74_years'] + \
                    data['rural_75-79_years'] + \
                    data['rural_80-84_years'] + \
                    data['rural_85-89_years'] + \
                    data['rural_90-94_years']
data = data.drop(['rural_50-54_years', 'rural_55-59_years', 'rural_60-64_years',
                  'rural_65-69_years', 'rural_70-74_years', 'rural_75-79_years',
                  'rural_80-84_years', 'rural_85-89_years', 'rural_90-94_years'], axis=1)

data
```

нормировка данных

```python
scaler_x = StandardScaler()
x = scaler_x.fit_transform(data[['volume_serv_medicine_2017', 'urban', 'urban_50+', 'density', 'rural_50+', 'humidity_median',
                               'volume_serv_transport_2017']])
scaler_y = StandardScaler()
y = scaler_y.fit_transform(data[['inf_rate']])
```

```python
model = KNeighborsRegressor(n_neighbors=3, weights='distance', p=1)
model.fit(x, y)
```

загрузка тестового датасета

```python
testset = pd.read_csv('covid_data_test.csv', index_col='Unnamed: 0')
testset
```

выделение собственных признаков

```python
testset['urban_50+'] = testset['urban_50-54_years'] + \
                    testset['urban_55-59_years'] + \
                    testset['urban_60-64_years'] + \
                    testset['urban_65-69_years'] + \
                    testset['urban_70-74_years'] + \
                    testset['urban_75-79_years'] + \
                    testset['urban_80-84_years'] + \
                    testset['urban_85-89_years'] + \
                    testset['urban_90-94_years']
testset = testset.drop(['urban_50-54_years', 'urban_55-59_years', 'urban_60-64_years',
                  'urban_65-69_years', 'urban_70-74_years', 'urban_75-79_years',
                  'urban_80-84_years', 'urban_85-89_years', 'urban_90-94_years'], axis=1)

testset['rural_50+'] = testset['rural_50-54_years'] + \
                    testset['rural_55-59_years'] + \
                    testset['rural_60-64_years'] + \
                    testset['rural_65-69_years'] + \
                    testset['rural_70-74_years'] + \
                    testset['rural_75-79_years'] + \
                    testset['rural_80-84_years'] + \
                    testset['rural_85-89_years'] + \
                    testset['rural_90-94_years']
testset = testset.drop(['rural_50-54_years', 'rural_55-59_years', 'rural_60-64_years',
                  'rural_65-69_years', 'rural_70-74_years', 'rural_75-79_years',
                  'rural_80-84_years', 'rural_85-89_years', 'rural_90-94_years'], axis=1)
```

нормировка данных

```python
x_t = scaler_x.fit_transform(testset[['volume_serv_medicine_2017', 'urban', 'urban_50+', 'density', 'rural_50+', 'humidity_median',
                                      'volume_serv_transport_2017']])
```

```python
testset['inf_rate'] = scaler_y.inverse_transform(model.predict(x_t))
```
