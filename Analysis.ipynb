грузим библиотеки

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

грузим датасет

```python
dataset = pd.read_csv('covid_data_train.csv')
for col in dataset.columns:
    print(col)
```

отбираем первичные признаки

```python
selected_columns = ['district',
                    'region_x',
                    'whole_population',
                    'urban',
                    'rural',
                    'density',
                    'avg_temp_median',
                    'humidity_median',
                    'urban_50-54_years',
                    'urban_55-59_years',
                    'urban_60-64_years',
                    'urban_65-69_years',
                    'urban_70-74_years',
                    'urban_75-79_years',
                    'urban_80-84_years',
                    'urban_85-89_years',
                    'urban_90-94_years',
                    'rural_50-54_years',
                    'rural_55-59_years',
                    'rural_60-64_years',
                    'rural_65-69_years',
                    'rural_70-74_years',
                    'rural_75-79_years',
                    'rural_80-84_years',
                    'rural_85-89_years',
                    'rural_90-94_years',
                    'bus_march_travel_18',
                    'bus_april_travel_18',
                    'shops_and_malls',
                    'public_transport',
                    'life_quality_place_rating',
                    'ecology',
                    'life_costs',
                    'cleanness',
                    'volume_serv_medicine_2017',
                    'volume_serv_hotels_2017',
                    'volume_serv_transport_2017',
                    'volume_serv_education_2017',
                    'inf_rate']
data = dataset[selected_columns]
data = data.dropna(axis=0)
data = data.drop_duplicates(ignore_index=True)
data
```

```python
data['urban_50+'] = data['urban_50-54_years'] + \
                    data['urban_55-59_years'] + \
                    data['urban_60-64_years'] + \
                    data['urban_65-69_years'] + \
                    data['urban_70-74_years'] + \
                    data['urban_75-79_years'] + \
                    data['urban_80-84_years'] + \
                    data['urban_85-89_years'] + \
                    data['urban_90-94_years']
data = data.drop(['urban_50-54_years', 'urban_55-59_years', 'urban_60-64_years',
                  'urban_65-69_years', 'urban_70-74_years', 'urban_75-79_years',
                  'urban_80-84_years', 'urban_85-89_years', 'urban_90-94_years'], axis=1)

data['rural_50+'] = data['rural_50-54_years'] + \
                    data['rural_55-59_years'] + \
                    data['rural_60-64_years'] + \
                    data['rural_65-69_years'] + \
                    data['rural_70-74_years'] + \
                    data['rural_75-79_years'] + \
                    data['rural_80-84_years'] + \
                    data['rural_85-89_years'] + \
                    data['rural_90-94_years']
data = data.drop(['rural_50-54_years', 'rural_55-59_years', 'rural_60-64_years',
                  'rural_65-69_years', 'rural_70-74_years', 'rural_75-79_years',
                  'rural_80-84_years', 'rural_85-89_years', 'rural_90-94_years'], axis=1)

data['bus_travel'] = (data['bus_march_travel_18'] + data['bus_april_travel_18']) / 2
data = data.drop(['bus_march_travel_18', 'bus_april_travel_18'], axis=1)

data
```

```python
heatmap_columns = ['volume_serv_medicine_2017', 'urban', 'urban_50+', 'density', 'rural_50+', 'humidity_median', 'volume_serv_transport_2017', 'inf_rate']
sns.heatmap(data[heatmap_columns].corr(method='pearson'), annot=True, fmt='0.2f')
```
